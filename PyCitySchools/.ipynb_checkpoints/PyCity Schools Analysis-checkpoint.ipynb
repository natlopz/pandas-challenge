# PyCity School Analysis

Based on the relationship between the spending budget per student and their academic outcomes, we expect to see that higher spending per student directly correlates with better academic scores on subjects tested. However, with this data set, this is not the case. According to the "Highest-Performing Schools (by % Overall Passing)" data set, Cabrera High School resulted in first place with a $582 spending budget per student and an overall passing score of 91.33%. Compared to the ranking school in the "Bottom Performing Schools (By % Overall Passing)" data set, Johnson High School resulted in a 650-dollar spending budget per student and scored an overall passing rate of 53.53%. These results indicate that how the school is spending their total budget brings more importance than how much they are spending. Furthermore, one can observe the impact of school size on student's academic performance and overall passing rates. Looking at the data we discussed above, Cabrera High School has a total student population of 1858 compared to Johnson Highschool with a total student population of 4761. Schools with fewer students may attribute to higher academic scores and passing rates where teachers can give more individual attention to students resulting in a student's higher chance of success. These observations highlighted in this analysis suggest that factors such as financial allocation and school size influence student's academic performance.

```python
# Import
import pandas as pd
from pathlib import Path
```

```python
# File Path
school_data = Path(r'C:\\Users\\lnata\\GitHub\\pandas-challenge\\Resources\\schools_complete.csv')
student_data = Path(r'C:\\Users\\lnata\\GitHub\\pandas-challenge\\Resources\\students_complete.csv')

# Read the .csv files
school_df = pd.read_csv(school_data)
student_df = pd.read_csv(student_data)

# Merge both .csv files into one
df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])

# Print
df.head()
```

## District Summary

```python
# Calculating the total number of unique schools
school_count = df["school_name"].nunique()
school_count
```

```python
# Calculating the total number of students
student_count = df["Student ID"].nunique()
student_count
```

```python
# Calculating total budget
t_budget = df.drop_duplicates(subset=["school_name"])["budget"].sum()
t_budget
```

```python
# Calculating average math score
avg_math_score = df["math_score"].mean()
avg_math_score
```

```python
# Calculating average reading score
avg_reading_score = df["reading_score"].mean()
avg_reading_score
```

```python
# Calculating the percentage of student who pass math! (Greater than or equal to 70)
passing_mscore = df[(df["math_score"] >= 70)].count()["student_name"]
passing_mpercent = passing_mscore / float(student_count) * 100
passing_mpercent
```

```python
# Caculating the percentage of students who pass reading! (Greater than or equal to 70)
passing_rscore = df[(df["reading_score"] >= 70)].count()["student_name"]
passing_rpercent = passing_rscore / float(student_count) * 100
passing_rpercent
```

```python
# Calculating the overall percentage of students who passed both math and reading! (Greater than or equal to 70)
passing_both = df[(df["math_score"] >= 70) & (df["reading_score"] >= 70)].count()["student_name"]
overall_rate = passing_both / float(student_count) * 100
overall_rate
```

```python
# Making a DataFrame that summarizes key metrics with data calculated
district_summary = pd.DataFrame({
    "Total Schools": [school_count],
    "Total Students": [student_count],
    "Total Budget": [t_budget],
    "Average Math Score": [avg_math_score],
    "Average Reading Score": [avg_reading_score],
    "% Passing Math": [passing_mpercent],
    "% Passing Reading": [passing_rpercent],
    "% Overall Passing": [overall_rate]
})

# Formatting values
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

district_summary
```

## School Summary

```python
# Select all school types
school_types = school_df.set_index("school_name")["type"]
school_types
```

```python
# Calculate the total student count per school
per_school_counts = df["school_name"].value_counts()
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget = school_df.set_index("school_name")["budget"]
per_school_capita = per_school_budget / per_school_counts
```

```python
# Calculate the average math test scores per school
per_school_math_agg = df.groupby("school_name").agg({"math_score": 'mean'})
per_school_math = per_school_math_agg["math_score"]

per_school_math
```

```python
# Calculate the average reading test scores per school
per_school_reading_agg = df.groupby("school_name").agg({"reading_score": 'mean'})
per_school_reading = per_school_reading_agg["reading_score"]

per_school_reading
```

```python
# Calculate the number of students per school with math scores of 70 or higher
students_passing_math = df[df["math_score"] >= 70]
school_students_passing_math = students_passing_math.groupby("school_name").count()["student_name"]

school_students_passing_math_percentage = (school_students_passing_math / per_school_counts) * 100
school_students_passing_math_percentage
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading = df[df["reading_score"] >= 70]
school_students_passing_reading = students_passing_reading.groupby("school_name").count()["student_name"]

school_students_passing_reading_percentage = (school_students_passing_reading / per_school_counts) * 100
school_students_passing_reading_percentage
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = df[
    (df["reading_score"] >= 70) & (df["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_students_passing_math / per_school_counts * 100
per_school_passing_reading = school_students_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100
overall_passing_rate
```

```python
# Create a DataFrame with the necessary columns
per_school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": per_school_counts,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Math Score": per_school_math,
    "Average Reading Score": per_school_reading,
    "% Passing Math": school_students_passing_math_percentage,
    "% Passing Reading": school_students_passing_reading_percentage,
    "% Overall Passing": overall_passing_rate
})

# Formatting budget columns for readability (optional)
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
print(per_school_summary)
```

## Highest-Performing Schools (by % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in descending order and display the top 5 rows.
top_schools = per_school_summary.sort_values("% Overall Passing", ascending=False)
top_schools.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

```python
# Sort the schools by `% Overall Passing` in ascending order and display the top 5 rows.
bottom_schools = bottom_schools = per_school_summary.sort_values("% Overall Passing", ascending=True)
bottom_schools.head(5)
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = df[(df["grade"] == "9th")]
tenth_graders = df[(df["grade"] == "10th")]
eleventh_graders = df[(df["grade"] == "11th")]
twelfth_graders = df[(df["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grader_math_scores = ninth_graders.groupby('school_name').math_score.mean()
tenth_grader_math_scores = tenth_graders.groupby('school_name').math_score.mean()
eleventh_grader_math_scores = eleventh_graders.groupby('school_name').math_score.mean()
twelfth_grader_math_scores = twelfth_graders.groupby('school_name').math_score.mean()

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = math_scores_by_grade = pd.DataFrame({
    "9th": ninth_grader_math_scores,
    "10th": tenth_grader_math_scores,
    "11th": eleventh_grader_math_scores,
    "12th": twelfth_grader_math_scores
})

# Minor data wrangling
math_scores_by_grade.index.name = None

# Print!
math_scores_by_grade
```

## Reading Score by Grade 

```python
# Use the code provided to separate the data by grade
ninth_graders = df[(df["grade"] == "9th")]
tenth_graders = df[(df["grade"] == "10th")]
eleventh_graders = df[(df["grade"] == "11th")]
twelfth_graders = df[(df["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grader_reading_scores = ninth_graders.groupby('school_name').reading_score.mean()
tenth_grader_reading_scores = tenth_graders.groupby('school_name').reading_score.mean()
eleventh_grader_reading_scores = eleventh_graders.groupby('school_name').reading_score.mean()
twelfth_grader_reading_scores = twelfth_graders.groupby('school_name').reading_score.mean()

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = reading_scores_by_grade = pd.DataFrame({
    "9th": ninth_grader_reading_scores,
    "10th": tenth_grader_reading_scores,
    "11th": eleventh_grader_reading_scores,
    "12th": twelfth_grader_reading_scores
})

# Minor data wrangling
reading_scores_by_grade.index.name = None

# Print!
reading_scores_by_grade
```

## Scores by School Spending

```python
per_school_summary["Per Student Budget"] = per_school_budget / per_school_counts
```

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()

# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_summary["Per Student Budget"], spending_bins, labels=labels)
school_spending_df
```

```python
# Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby("Spending Ranges (Per Student)", observed=True)["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby("Spending Ranges (Per Student)", observed=True)["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby("Spending Ranges (Per Student)", observed=True)["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby("Spending Ranges (Per Student)", observed=True)["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby("Spending Ranges (Per Student)", observed=True)["% Overall Passing"].mean()

# Assemble into DataFrame
spending_summary = pd.DataFrame({
  "Average Math Score": spending_math_scores,
  "Average Reading Score": spending_reading_scores,
  "% Passing Math": spending_passing_math,
  "% Passing Reading": spending_passing_reading,
  "% Overall Passing": overall_passing_spending
})

# Display results
spending_summary
```

## Scores by School Size

```python
per_school_summary["Per Student Budget"] = per_school_budget / per_school_counts
```

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(per_school_summary["Total Students"], size_bins, labels=labels)
per_school_summary
```

```python
per_school_summary['School Size'] = per_school_summary['School Size'].astype('category')

# Calculate averages for the desired columns. 
size_math_scores = per_school_summary.groupby("School Size", observed=True)["Average Math Score"].mean()
size_reading_scores = per_school_summary.groupby("School Size", observed=True)["Average Reading Score"].mean()
size_passing_math = per_school_summary.groupby("School Size", observed=True)["% Passing Math"].mean()
size_passing_reading = per_school_summary.groupby("School Size", observed=True)["% Passing Reading"].mean()
size_overall_passing = per_school_summary.groupby("School Size", observed=True)["% Overall Passing"].mean()
```

```python
# Create a DataFrame called `size_summary` that breaks down school performance based on school size (small, medium, or large).
# Use the scores above to create a new DataFrame called `size_summary`
size_summary = pd.DataFrame({
  "Average Math Score": size_math_scores,
  "Average Reading Score": size_reading_scores,
  "% Passing Math": size_passing_math,
  "% Passing Reading": size_passing_reading,
  "% Overall Passing": size_overall_passing
})

# Display results
size_summary
```

## Scores by School Type

```python
# Group the per_school_summary DataFrame by "School Type" and average the results.
average_math_score_by_type = per_school_summary.groupby(["School Type"])["Average Math Score"].mean()
average_reading_score_by_type = per_school_summary.groupby(["School Type"])["Average Reading Score"].mean()
average_percent_passing_math_by_type = per_school_summary.groupby(["School Type"])["% Passing Math"].mean()
average_percent_passing_reading_by_type = per_school_summary.groupby(["School Type"])["% Passing Reading"].mean()
average_percent_overall_passing_by_type = per_school_summary.groupby(["School Type"])["% Overall Passing"].mean()
```
